## Helper Functions

```python
def render_mpl_table(data, col_width=5.0, row_height=0.625, font_size=14,
                     header_color='#ff1a75', row_colors=['#f1f1f2', 'w'], edge_color='w',
                     bbox=[0, 0, 1, 1], header_columns=0,
                     ax=None, **kwargs):
    if ax is None:
        size = (np.array(data.shape[::-1]) + np.array([0, 1])) * np.array([col_width, row_height])
        fig, ax = plt.subplots(figsize=size)
        ax.axis('off')
    mpl_table = ax.table(cellText=data.values, bbox=bbox, colLabels=data.columns, **kwargs)
    mpl_table.auto_set_font_size(False)
    mpl_table.set_fontsize(font_size)

    for k, cell in mpl_table._cells.items():
        cell.set_edgecolor(edge_color)
        if k[0] == 0 or k[1] < header_columns:
            cell.set_text_props(weight='bold', color='w',fontsize=14, fontfamily='serif')
            cell.set_facecolor(header_color)
        else:
            cell.set_facecolor(row_colors[k[0]%len(row_colors)])
    return ax.get_figure(), ax

def mean_absolute_percentage_error(y, yhat):
    return np.mean(np.abs((y - yhat)/y))


def ml_error(model_name,y,yhat):
    mae = mean_absolute_error(y, yhat)
    mape = mean_absolute_percentage_error(y,yhat)
    rmse = np.sqrt(mean_squared_error(y,yhat))
    
    return pd.DataFrame({'Model Name': model_name,
                         'MAE': mae,
                         'MAPE': mape,
                         'RMSE': rmse}, index = [0])
```

## Change Types & Data Cleaning

We neet to:

- **1**: remove columns from the data set that are not useful for price prediction
- **2**: remove columns that has very high percentage of na values
- **3**: transform price and other numeric columns from string to int,float type 

```python
# last_scraped -> datetime
df1['last_scraped'] = pd.to_datetime(df1['last_scraped'], format = "%Y-%m-%d")

# host_since -> datetime
df1['host_since'] = pd.to_datetime(df1['host_since'], format = "%Y-%m-%d")

# first_review -> to datetime
df1['first_review'] = pd.to_datetime(df1['first_review'], format = "%Y-%m-%d")

# last_review -> to datetime
df1['last_review'] = pd.to_datetime(df1['last_review'], format = "%Y-%m-%d")

# host_response_time -> to boolean
# ======= within an hour = 0 | within a few hours = 1 | within a day = 2 | a few days or more = 3 =======
df1['host_response_time'] = df1['host_response_time'].apply(lambda x: 0 if x == 'within an hour'
                                                                 else 1 if x == 'within a few hours'
                                                                 else 2 if x == 'within a day' else 3)

# host_response_rate -> strip %
df1['host_response_rate'] = df1['host_response_rate'].str.replace("%", "")
df1['host_response_rate'] = pd.to_numeric(df1['host_response_rate'])

# host_acceptance_rate -> strip %
df1['host_acceptance_rate'] = df1['host_acceptance_rate'].str.replace("%", "")
df1['host_acceptance_rate'] = pd.to_numeric(df1['host_acceptance_rate'])

#host_has_profile_pic -> To boolean
df1['host_has_profile_pic'] = df1['host_has_profile_pic'].apply(lambda x: 1 if x == 't' else 0) 
df1['host_has_profile_pic'] = pd.to_numeric(df1['host_has_profile_pic'])

# host_is_superhost -> to boolean
df1['host_is_superhost'] = df1['host_is_superhost'].apply(lambda x: 1 if x == 't' else 0)

# host_identity_verified -> to boolean
df1['host_identity_verified'] = df1['host_identity_verified'].apply(lambda x: 1 if x == 't' else 0)

# has_availability -> to boolean
df1['has_availability'] = df1['has_availability'].apply(lambda x: 1 if x == 't' else 0)

# price -> strip $
df1['price'] = df1['price'].str.replace("$","", regex = True)
df1['price'] = df1['price'].str.replace(",", "", regex = True)
df1['price'] = pd.to_numeric(df1['price'])
```

In this section, we will finally build our predictive models. Therefore, we will use 5 machine learning algorithms, which will be:

- **Linear Regression**
- **Lasso**
- **Random Forest Regressor**
- **LGBM**

For each algorithm, we will build what we call the cross-validation technique. Cross-validation is primarily used in applied machine learning to estimate the skill of a machine learning model on unseen data. That is, to use a limited sample in order to estimate how the model is expected to perform in general when used to make predictions on data not used during the training of the model.

It is a popular method because it is simple to understand and because it generally results in a less biased or less optimistic estimate of the model skill than other methods, such as a simple train / test split.
